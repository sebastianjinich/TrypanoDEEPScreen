```python
from utils import gat_tanimoto_similarity_triangle

gsk_similarity = gat_tanimoto_similarity_triangle(gsk)
```

```python
duplicated_compunds_bioactivity_conflict = gsk[gsk.duplicated("comp_id",keep=False)]


def look_very_smiliars(comp_id,similarity_triangle,whole_df,min_tanimoto = 0.5, min_mw_ratio = 0.8):

    tanimotos = similarity_triangle.loc[comp_id,:]
    high_tanimotos = tanimotos[tanimotos>min_tanimoto].index.to_list()

    if len(high_tanimotos)>0:
        
        similar_comps_similar_mw = list()

        for comp_to_check in high_tanimotos:
            molecular_weights = whole_df.loc[whole_df["comp_id"].isin((comp_to_check,comp_id)),"mw_freebase"]
            ratio = min(molecular_weights)/max(molecular_weights)
            if ratio > min_mw_ratio:
                similar_comps_similar_mw.append(comp_to_check)
    
        return similar_comps_similar_mw
    
    else:
        return pd.NA


duplicated_compunds_bioactivity_conflict.loc[:,"very_similar"] = duplicated_compunds_bioactivity_conflict.apply(lambda x : look_very_smiliars(x["comp_id"], similarity_triangle = gsk_similarity, whole_df = gsk),axis=1)
```

```python
comps_to_drop = list()

for comps_list in duplicated_compunds_bioactivity_conflict["very_similar"]:
    comps_to_drop += comps_list

comps_to_drop = pd.Series(comps_to_drop).unique()
```

```python
gsk = gsk[~gsk.comp_id.isin(comps_to_drop)]
```
