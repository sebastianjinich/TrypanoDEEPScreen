```python
import sys; sys.path.insert(0, '..'); sys.path.insert(0, '../utils')
from utils.baseline import fingerprint_generator_ASP
from rdkit import DataStructs
```

```python
train_1 = fingerprint_generator_ASP(train.reset_index(drop=True))
validation = fingerprint_generator_ASP(validation.reset_index(drop=True))
test = fingerprint_generator_ASP(test.reset_index(drop=True))
```

```python
def calc_similairity_features(fp,train_fps):
    similarity_features = [DataStructs.BraunBlanquetSimilarity(fp,fp_train) for fp_train in train_fps]
    return similarity_features

train_fps = train_1.loc[:,"fps"]
validation.loc[:,"features"] = validation.loc[:,"fps"].apply(lambda x: calc_similairity_features(x,train_fps))
test.loc[:,"features"] = test.loc[:,"fps"].apply(lambda x: calc_similairity_features(x,train_fps))
```

```python
import numpy as np

from scipy.stats import uniform, randint

from sklearn.datasets import load_breast_cancer, load_diabetes, load_wine
from sklearn.metrics import auc, accuracy_score, confusion_matrix, roc_auc_score
from sklearn.model_selection import cross_val_score, GridSearchCV, KFold, RandomizedSearchCV, train_test_split

import xgboost as xgb
```

```python
X = validation.features
X = np.array(X.tolist())
y = validation.bioactivity.to_numpy()

xgb_model = xgb.XGBClassifier(objective="binary:logistic", random_state=42, eval_metric = "auc", colsample_bytree = 0.7711680375207983, gamma=0.05089123631020187, learning_rate=0.07585774175529961, max_depth=4, n_estimators=147, subsample=0.6642725493038223)
xgb_model.fit(X, y)

test_features = np.array(test.features.tolist())
y_proba = xgb_model.predict_proba(test_features)[:, 1] 

print("auroc", roc_auc_score(test.bioactivity,y_proba))
```
